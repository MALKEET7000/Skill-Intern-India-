```python
#importing libraries
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns
```

```python
#loading dataset
df = pd.read_csv("SampleSuperstore.csv")
df.head()
```

```python
df.shape
```

```python
df.info
```

```python
df.describe()
```

```python
# finding null values
df.isnull().sum()
```

```python
# Finding duplicate values from dataset
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace = True)
df
```

```python
#Restting index
df.reset_index(inplace = True)
df
```

```python
# dropping column index
df.drop(columns=['index'],inplace = True)
df
```

```python
df.describe()
```

```python
# Check correlation
df.corr()
```

```python
# Check correlation by heatmap
fig, axes = plt.subplots(1,1, figsize = (10,5))
sns.heatmap(df.corr(), annot= True)
plt.show()
```

we can conclude from heat map  
Brightest color shows the strong correlation,
Dark color shows no correlation,

Above Heatmap shows

Sales and Profit are Moderately Correlated,
Discount and Profit are Negatively Correlated,
Quantity and Profit are less Moderately Correlated,
It seems Profit is correlated with Sales.

```python
#scatter plot
plt.scatter(df['Sales'],df['Profit'], c ='RED' )
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Sales v/s Profit')
plt.show()
```

when the sales are high, a company mostly gets profit.
At lower sales, a company either gets profit or loss

```python
sns.countplot(x='Quantity',hue='Region',data=df )
```

above count plot shows the regionwise quantity count.

```python
fig, axes = plt.subplots(1,1)
grouped = df.groupby('Region')['Sales'].sum()
grouped.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.title("Pie Chart showing combined Regionwise sales")
plt.legend()
plt.show()
grouped = df.groupby('Region')['Profit'].sum()
grouped.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.title("Pie Chart showing combined Regionwise profit")
plt.legend()
plt.show()
grouped = df.groupby('Region')['Discount'].sum()
grouped.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.title("Pie Chart showing combined Regionwise Discount")
plt.legend()
plt.show()
```

above three pie charts shows combined regionwise sales,profit and discount.

```python
grouped = df.groupby('Category')['Sales'].sum()
grouped.plot(kind='bar',color='blue')
plt.title("Bar Chart showing sales in different categories")
plt.xlabel("Category")
plt.ylabel("Sum of Sales")
plt.show()
```

```python
grouped = df.groupby('Category')['Discount'].sum()
grouped.plot(kind='bar',color='Red')
plt.title("Bar Chart showing sales in different categories")
plt.xlabel("Category")
plt.ylabel("Discount")
plt.show()
```

# CONCLUSION.
Correlation Matrix showing positive relationship between Profit & Sales, Quantity & Sales.
Company offering more Discount on Office Sppliers & furniture Category.
In the Central Region company(Store) giving lots of discount but the Output is very less in quantity.
The mode of shipping doesn't affect much to the sales.
