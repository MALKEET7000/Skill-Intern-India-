# Prediction using Supervised ML 

# Predict the percentage of an student based on the no. of study hours.

# Using Linear Regression

```python
# importing the important libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading our data set.
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
# showing first 7 row data 
data.head(7)
```

```python
#Finding outliers in data set using boxplot
```

```python
plt.subplot(1,2,1)
sns.boxplot(data['Hours'])
plt.subplot(1,2,2)
sns.boxplot(data['Scores'])
```

```python
#from the above we can find that the dataset has no outliers.
```

```python
# using scatter plot we find the relationship between the the two variables.
sns.scatterplot(data=data, x='Hours', y='Scores',color='green', marker='^', s=100)
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Scatter Plot showing the relationship between the two variables Hours & Scores')
plt.show()
```

```python
# dividing the dataset into dependent and independent variables using iloc function.
```

```python
X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values  
```

```python
# using train_test_split to split the data by importing sklearn library.
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=0) 
```

```python
# now we train the dataset
```

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 
```

```python
# Plotting the regression line
sns.regplot(data=data, x='Hours', y='Scores', ci=None)  # ci=None disables confidence intervals
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Scatter Plot with Intercept Line between the two variables')
plt.show()
```

```python
# Testing data - In Hours
print(X_test)
# Predicting the scores
y_pred = regressor.predict(X_test)
```

```python
# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```

```python
# test with given data 9.25 hours.
hours = np.array([9.25])
hours = hours.astype('float64')
hours = hours.reshape(-1, 1)
own_pred = regressor.predict(hours)
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```
